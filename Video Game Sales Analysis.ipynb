```python
#import stats from the scipy library
from scipy import stats
```

```python
#calculate the 5% trimmed mean
stats.trim_mean(df["Global_Sales"], 0.05)
```

The trim_mean function, cuts out a percentage of values (the upper and lower 5% of values in this case), that are at both ends of our distributions. The returns a mean without potential outliers, forces our variable to be symmetrical.

The 5% trimmed mean is 0.32 - which is less than the normal mean calculation, and further suggests that there may be high outliers which are bringing the mean up.

## Outliers

```python

# check for outliers
q1 = df['Global_Sales'].quantile(0.25)
q3 = df['Global_Sales'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
print(df[(df['Global_Sales'] < lower_bound) | (df['Global_Sales'] > upper_bound)].count())
```

The above code suggests that there are several outlier in the global sales variable. 

As the graphs indicated above, these are all high outliers. 

It was decided not to remove any outliers, as considering the domain, it is highly unlikely that the outliers are due to random error. 

Outliers should not be removed when they are not due to data collection error, or when they are numerous (Chirag Goyal, 2021).

#**Third section: Linear Regression Model**

In this script, we first import the necessary libraries, including NumPy and the linear regression model from scikit-learn. Next, we define the X and Y variables and create a linear regression model. Then, we fit the model to the data using the fit method.

After fitting the model, we print the model coefficients, intercept, and R-squared value. The coefficients represent the slope of the regression line, and the intercept is the point where the regression line crosses the y-axis. The R-squared value, also known as the coefficient of determination, is a measure of how well the regression line fits the data. A value of 1 indicates a perfect fit, while a value of 0 indicates that the line does not explain any of the variance in the data.

These values can be accessed using the coef_, intercept_, and score methods of the linear regression model. By printing these values, we can interpret the results of the model and better understand the relationship between the X and Y variables.

```python
# Import the necessary libraries
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array(df['EU_Sales']).reshape(-1,1)
Y = np.array(df['NA_Sales'])

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Use the model to make predictions
predictions = model.predict(X)

# Print the predictions
print(predictions)
```

```python
# Print the model coefficients
print("Coefficients: ", model.coef_)

# Print the model intercept
print("Intercept: ", model.intercept_)

# Print the model R-squared value
print("R-squared: ", model.score(X, Y))
```

The R-Squared coefficient indicates that 58.57% of the variance in NA_Sales is due to EU_Sales.
